## Clustering Japanese and French restaurant using KMean

```python
X_fr =df_fr1.values[:,1:]
#X =df_fr1.values
X_fr
```

```python
clusterNum = 4
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X_fr)
labels = k_means.labels_
print(labels)
```

```python
df_fr1["Clus_km"] = labels
df_fr1.head(10)
```

```python
# create map
map_houston_clus = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(clusterNum+2)
ys = [i + x + (i*x)**2 for i in range(clusterNum+2)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, cluster in zip(df_fr1['lat'], df_fr1['lng'], df_fr1['Clus_km']):
 #   label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.RegularPolygonMarker(
        [lat, lon],
        number_of_sides=3,
        radius=10,
 #       popup=label,
        color=rainbow[cluster-1],
#        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_houston_clus)
       
map_houston_clus
```
